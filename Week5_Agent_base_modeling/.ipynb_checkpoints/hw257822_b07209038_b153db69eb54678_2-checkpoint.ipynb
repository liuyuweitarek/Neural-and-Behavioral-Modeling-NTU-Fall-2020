## 1 Nonlinear love triangle (5 points)

Please use Euler integration with a small step size (say dt=0.01) and the suggested parameters (a,b,c,d,e,f)=(-3,4,-7,2,2,-1) to simulate Eq.(5) in the following paper:

Sprott, J. C. (2004). Dynamical models of love. Nonlinear dynamics, psychology, and life sciences, 8(3), 303-314.

### 1.1  Coding the simulation (3 points)

Please use differnt colors/shapes to put Rj(t), J, Rg(t), & G(t) curves in the same plot.

```python
a,b,c,d,e,f = -3,4,-7,2,2,-1 # Model parameters

rj, rg, j, g = [0.1], [0.1], [0.1], [0.1] # Initial values
dt = .001 # step size for numerical integration
t = arange(0,30,dt) # integration period

# ---- Euler Integration ---- #
for i in t:
    rj1, rg1, j1, g1 = rj[-1], rg[-1], j[-1], g[-1] # the last element
     
    rj2 = rj1 + dt*(a*rj1 + b*(j1-g1)*(1-abs(j1-g1))) # Euler integration
    rg2 = rg1 + dt*(a*rg1 + b*(g1-j1)*(1-abs(g1-j1))) # Euler integration
    j2  = j1  + dt*(c*rj1*(1-abs(rj1)) + d*j1)        # Euler integration
    g2  = g1  + dt*(e*rg1*(1-abs(rg1)) + f*g1)        # Euler integration
    
    rj.append(rj2)
    rg.append(rg2)
    j.append(j2)
    g.append(g2)

# ---- Euler Integration Done ---- #

# ---- Plot ---- #
figure(num=None, figsize=[10,5])
subplot(1,2,1)
plot(t,rj[0:-1],'b') # Romeo to Juliet
plot(t,rg[0:-1],'g') # Romeo to Guinevere
plot(t,j[0:-1],'m')  # Juliet
plot(t,g[0:-1],'r')  # Guinevere
legend(['Romeo to Juliet','Romeo to Guinevere','Juliet','Guinevere']);
title('Time Series')
xlabel('time')
ylabel('love')

subplot(1,2,2)
a,b,c,d,e,f = -3,4,-7,2,2,-1 # Model parameters
rj_s, j_s = meshgrid(arange(-4,3,0.01), arange(-4,3,0.01))
g1 = 0
rj_dot = a*rj_s + b*(j_s-g1)*(1-abs(j_s-g1))
j_dot = c*rj_s*(1-abs(rj_s)) + d*j_s
# streamplot(rj_s, j_s, rj_dot, j_dot) # 這裡不能畫二維streamline，因為還有g、Rg的變數，這裡只能畫固定g、R下的流場，顯然畫出來會不平行橘線，因為橘線上g、rg一直在變，要畫會變成四維空間。
plot(rj[-1],j[-1], 'r^')
print(rj[-1],j[-1])
title('Streamline of Rj and J')
xlabel('J')
ylabel('Rj')

plot(rj, j)
show()
```

隨者G由討厭到喜歡，RJ和J的關係從J很討厭R到J超喜歡R，反正就是你喜歡我也要喜歡。R則變化幅度不大。

## 2 Fix the simulation of "Growth and underinvestment" (5 points)

Although the simulation and explanation of "underinvested growth" seem reasonable, the simulation doesn't appear to describe "sufficiently invested growth" where the performance varialbe should be able to increase over time. Please do whatever you can to have a model that can simulate both phenomena. 

```python
# --- Original Code--- #
performance_standard=100 # Model parameter
Capacity,Demand=[80],[1] # Initial values
dt=.1 # step size for numerical integration
t=arange(0,13,dt) # integration period
perf,inves=[],[]
for i in t:
    Capacity1,Demand1=Capacity[-1],Demand[-1] # the last element
    growing_action=0.01*Demand1
    performance=Capacity1-Demand1
    change_rate=0.75*growing_action*performance
    perf.append(performance)
    growth_and_underinverstment=performance_standard-performance
    investment=0.01*growth_and_underinverstment # !!!!!!!! Change 0.01 to 10 doesn't help !!!!!!!!
    inves.append(investment)
    Capacity2=Capacity1+dt*investment # Euler integration
    Demand2=Demand1+dt*change_rate # Euler integration
    Capacity.append(Capacity2)
    Demand.append(Demand2)
plot(t,Capacity[0:-1],'bv')
plot(t,inves,'ro')
plot(t,perf,'g.')
plot(t, Demand[0:-1])
legend(['Capacity','Investment','Performance','Demand']);
```

### 一、修改想法：
這個投資不足的模型中有幾個重要的關係必須釐清
#### | 1. 時間延滯：投資必須經過一段時間才能反應在產能上。
時間延滯將導致投資來不及拯救快下降的績效。
#### | 2. Capacity與Demand的成長速度：相對速度影響Performance走向。
可以進行多項式次方的討論，如果後者次方高於前者，時間增加，Demand一定會後來居上。因此要盡可能讓Capacity的次方高於Demand次方。
#### | 3. 參考第五項試煉書中此系統基模的管理方針
用比較生活化的角度去思考，怎麼樣才能提早發現Demand未來的快速成長，以達到在Performance下降之前，準備好足夠的Capacity去應對激增的Demand。

### 二、觀察原始模型
希望修正的點：
1. Demand的成長速度明顯高於Capacity
2. Investment的成長速度需要增加
3. Performance趨勢不應該下降這麼劇烈，處理完1應該就可以同時達到3。

## 最終模型

```python
# --- 修改目標：投資和表現都有成長，但是投資明顯比成長多 --- #
performance_standard=100 # Model parameter
Capacity,Demand,inves=[80],[1],[0] # Initial values
dt=.1 # step size for numerical integration
t=arange(0,15,dt) # integration period
perf=[]

# --- Euler Integration ---#
for i in t:
    # the last element
    Capacity1, Demand1, inves1 = Capacity[-1], Demand[-1],inves[-1]  # 修改一
    growing_action = 0.01*Demand1  # 成長的行動
    
    
    # --- 績效 --- #
    performance = Capacity1 - Demand1
    change_rate = 0.4*growing_action * performance # 修改二  # 成長的行動
    perf.append(performance)
    growth_and_underinverstment = performance_standard - performance
    
    # --- 投資 --- #
    investment = 0.01*growing_action * growth_and_underinverstment * growth_and_underinverstment * growth_and_underinverstment # 修改三 # !!!!!!!! Change 0.01 to 10 doesn't help !!!!!!!!
    inves.append(investment) # 產能的投資
    
    
    # --- Euler Integration --- #
    # --- 1/產能提高 ： 投資正回饋、時間延滯
    # --- 2/需求提高 ： 成長行動正回饋、績效負回饋 
    Capacity2 = Capacity1 + dt*inves1 # 修改四  # Euler integration # 修改四
    Demand2   = Demand1 + dt*change_rate # Euler integration
    Capacity.append(Capacity2)
    Demand.append(Demand2)

# --- Integration Done ---#

# --- Plot --- #
plot(t,Capacity[0:-1],'bv')
plot(t,inves[0:-1],'ro')
plot(t,perf,'go')
plot(t, Demand[0:-1],'co')
legend(['Capacity','Investment','Performance','Demand']);
#print(perf[-10:-1])

# --- Plot Done ---#
```

### 三、修改詳述
> 修改一及修改四：加入投資對產能提高的時間延滯

將investment儲存，這個迴圈內先不能更新到Capacity，而在下個時間才更新。此舉是為了貼近人類看到需要提高產能才增加投資，投資並不能及時反應在產能。

> 修改二：調整chage_rate係數，0.75偏大，0.4較佳，可以使系統整體成長在可以觀察的範圍。

> 修改三：提高Investment的Order。

從計算Demand的算式發現 Change_rate = Demand*performance，約兩次方
但原始Investment只有 Investment = growth_and_underinvestment，約一次方
所以我提高Investment的growth_and_underinvestment變growth_and_underinvestment^3，約三次方

> 修改三：Investment考慮到Demand成長速度。

要達成「一、修改想法 第3點」，代表Investment需要考慮Demand的導數。這裡就直接多呈上一個growing_action，有效的提高了Investment對Demand的敏感度。

### 四、其他異常的模型
這些都是過程中的想法，供參考。

```python
# --- 修改目標：投資和表現都有成長，但是投資明顯比成長多 --- #
performance_standard=100 # Model parameter # 績效標準
Capacity,Demand=[80],[1] # Initial values
dt=.1 # step size for numerical integration
t=arange(0,13,dt) # integration period
perf,inves=[],[]

# --- Euler Integration ---#
for i in t:
    # the last element
    Capacity1, Demand1 = Capacity[-1], Demand[-1]

    
    # --- 績效 --- #
    performance = Capacity1 - Demand1 # 績效
    perf.append(performance)
    
    # --- 投資 --- #
    growth_and_underinverstment = 1 - performance/performance_standard  # 投資的認知需求
    investment = 100 * growth_and_underinverstment # 產能的投資  # !!!!!!!! Change 0.01 to 10 doesn't help !!!!!!!!
    inves.append(investment)

    # --- 需求提高 --- # 
    # 成長行動正回饋、績效負回饋
    growing_action = 0.2*Demand1 # 成長的行動
    change_rate =  performance/performance_standard # 成長的行動

    
    # --- Euler Integration --- #
    Capacity2 = Capacity1 + dt*investment # 產能      # Euler integration
    Demand2   = Demand1 + dt*(change_rate+growing_action)  # 成長的行動 # Euler integration
    Capacity.append(Capacity2)
    Demand.append(Demand2)

# --- Integration Done ---#

# --- Plot --- #
plot(t,Capacity[0:-1],'bv')
plot(t,inves,'ro')
plot(t,perf,'g.')
plot(t,Demand[0:-1])
legend(['Capacity','Investment','Performance']);
```

說明：將Performence標準化、將chance rate中「growing_action」和「performance」從原先的相乘關係變成線性組合關係。
